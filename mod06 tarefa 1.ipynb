#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_sinasc = pd.read_csv('sinasc_RO_2019.csv')
df_sinasc
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
%matplotlib notebook

fig = plt.figure()

graf1 = fig.add_subplot(2,3,1)
graf2 = fig.add_subplot(2,3,2)
graf3 = fig.add_subplot(2,3,3)
graf4 = fig.add_subplot(2,3,4)
graf5 = fig.add_subplot(2,3,5)

graf1.hist(df_sinasc['APGAR1'].dropna())
graf2.hist(df_sinasc['APGAR5'].dropna())
graf3.hist(df_sinasc['SEXO'].dropna())
graf4.hist(df_sinasc['GESTACAO'].dropna())
graf5.hist(df_sinasc['ESTCIVMAE'].dropna())
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5', y='CONSULTAS', estimator='mean', data=df_sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, ax = plt.subplots()
plt.close('all')
plt.rc('figure', figsize=(20, 18))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax=axes[0, 0], x='SEXO', data=df_sinasc)
sns.countplot(ax=axes[0, 1], x='GESTACAO', data=df_sinasc)
sns.countplot(ax=axes[1, 0], x='ESTCIVMAE', data=df_sinasc)
sns.countplot(ax=axes[1, 1], x='CONSULTAS', data=df_sinasc)
```
